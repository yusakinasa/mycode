```python
import torch

x = torch.arange(4.0)
x
```

```python
x.requires_grad_(True)
x.grad
```

```python
y = 2 * torch.dot(x, x)
y
```

```python
y.backward()
x.grad
```

通过调用反向传播函数来自动计算y关于x每个分量的梯度

```python
x.grad == 4 * x
```

```python
x.grad.zero_()
y = x.sum()
y.backward()
x.grad
```

深度学习中 ，我们的目的不是计算微分矩阵，而是批量中每个样本单独计算的偏导数之和

```python
x.grad.zero_()
y = x * x
y.sum().backward()
x.grad
```

将某些计算移动到记录的计算图之外

```python
x.grad.zero_()
y = x * x
u = y.detach()
z = u * x

z.sum().backward()
x.grad == u
```

```python
x.grad.zero_()
y.sum().backward()
x.grad == 2 * x
```

```python
def f(a):
    b = a * 2
    while b.norm() < 1000:
        b = b * 2
    if b.sum() > 0:
        c = b
    else:
        c = 100 * b
    return c

a = torch.randn(size=(), requires_grad=True)
d = f(a)
d.backward()

a.grad == d / a
```

即使构建函数的计算图需要通过Python控制流（例如，条件、循环或任意函数调用），我们仍然可以计算得到的变量的梯度
